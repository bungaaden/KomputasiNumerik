#Kasus: Tumbukan Dua Benda dengan Kerugian Energi

```
import numpy as np
import matplotlib.pyplot as plt
```

Parameter

```
m1 = 1.0  # massa benda 1 (kg)
m2 = 2.0  # massa benda 2 (kg)
v1i = np.array([3.0, 0.0])  # kecepatan awal benda 3 (m/s)
v2i = np.array([-1.0, 0.0])  # kecepatan awal benda -1 (m/s)
epsilon = 0.8  # koefisien restitusi
```

Simulasi dengan komputasi numerik

```
dt = 0.01  # langkah waktu (s)
num_steps = 1000  # jumlah langkah waktu
t = np.linspace(0, dt*num_steps, num_steps+1)
```

Inisialisasi array untuk kecepatan

```
v1 = np.zeros((num_steps+1, 2))
v2 = np.zeros((num_steps+1, 2))
v1[0] = v1i
v2[0] = v2i
```

Metode Euler untuk simulasi

```
for i in range(num_steps):
    # Hukum kekekalan momentum
    v1[i+1] = v1[i] + (m2/(m1+m2)) * (1 + epsilon) * np.dot(v2[i] - v1[i], v2[i] - v1[i]) / np.dot(v1[i] - v2[i], v1[i] - v2[i]) * (v2[i] - v1[i]) * dt
    v2[i+1] = v2[i] - (m1/(m1+m2)) * (1 + epsilon) * np.dot(v1[i] - v2[i], v1[i] - v2[i]) / np.dot(v1[i] - v2[i], v1[i] - v2[i]) * (v2[i] - v1[i]) * dt
```

Plot hasil simulasi

```
plt.plot(t, v1[:, 0], label='v1x')
plt.plot(t, v1[:, 1], label='v1y')
plt.plot(t, v2[:, 0], label='v2x')
plt.plot(t, v2[:, 1], label='v2y')
plt.xlabel('Waktu (s)')
plt.ylabel('Kecepatan (m/s)')
plt.title('Simulasi Tumbukan Dua Benda dengan Kerugian Energi')
plt.legend()
plt.grid(True)
plt.show()
```
